```python
import pandas as pd

df = pd.read_csv("neural_data.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df["latency"])
plt.title("Latency Boxplot")
plt.show()
```

```python
import seaborn as sns

sns.histplot(df["latency"], kde=True)
plt.title("Latency Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.scatter(df["noise_level"], df["latency"])
plt.xlabel("Noise Level")
plt.ylabel("Latency")
plt.title("Noise vs Latency")
plt.show()
```

## Data Analysis Insights
1. Latency Distribution

The latency values follow an approximately normal (bell-shaped) distribution centered around 1.0. Most neural responses fall within a moderate range, indicating stable system behavior with few extreme values.

2. Presence of Outliers

The boxplot shows a few outliers in latency values. These extreme values may represent unusual neural responses or measurement noise, but overall they are limited in number.

3. Strong Impact of Noise Level on Latency

The correlation matrix shows a strong positive correlation (0.70) between noise level and latency. This indicates that higher signal noise significantly increases neural response delay.
This suggests that reducing signal noise could improve prosthetic response time.

4. Moderate Effect of Electrode Distance

Electrode distance shows a moderate positive correlation (0.36) with latency. This suggests that larger distances between electrodes and neural sources increase response delay.

5. Weak Relationship of Other Features

Amplitude, frequency, and stimulus intensity show weak correlations with latency. This indicates they have limited influence on neural response timing in this dataset.

6. Visual Relationship Between Noise and Latency

The scatter plot shows a clear upward trend between noise level and latency, confirming that increasing noise leads to higher latency.


### Practical Interpretation
The analysis suggests that neural latency in prosthetic control systems is primarily affected by signal noise and electrode positioning. Optimizing signal quality and electrode placement may significantly improve response speed and system efficiency.
